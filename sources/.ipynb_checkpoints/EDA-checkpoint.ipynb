```python
import pandas as pd
import matplotlib.pyplot as plt

from tweet2pdf import api, get_tweets
```

```python
#select accounts and load data into df
df = pd.DataFrame()
for i in ['wojespn','ShamsCharania']:
    df = df.append(get_tweets(api = api, screen_name = i),ignore_index= True)
```

```python
df['date_time'] = pd.to_datetime(df['date'])
df['date'] = pd.to_datetime(df['date_time']).dt.date
```

```python
plot_data = df.groupby(by = ['date','name'],as_index = False).count()[['date','name','text']].rename(columns = {'text':'count'})
fig = plt.figure(figsize=(11,8))
ax1 = fig.add_subplot(111)

ax1.plot(plot_data['date'][plot_data['name'] == 'wojespn'], plot_data['count'][plot_data['name'] == 'wojespn'], label='Woj', color='c', marker='o')
ax1.plot(plot_data['date'][plot_data['name'] == 'ShamsCharania'], plot_data['count'][plot_data['name'] == 'ShamsCharania'], label = 'Shams', color='g', marker='o')

handles, labels = ax1.get_legend_handles_labels()
lgd = ax1.legend(handles, labels, loc='upper center', bbox_to_anchor=(1.15,1))
```
